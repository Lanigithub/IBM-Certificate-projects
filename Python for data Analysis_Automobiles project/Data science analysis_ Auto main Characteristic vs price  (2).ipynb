```python
# 4. Basics of Grouping
# The "groupby" method groups data by different categories. 
# The data is grouped based on one or several variables and analysis is performed on the individual groups.

# For example, let's group by the variable "drive-wheels". We see that there are 3 different categories of drive wheels.
```

```python
df['drive-wheels'].unique()  # function unique()
```

```python
# numpy.unique. Advertisements. This function returns an array of unique elements in the input array. 
# The function can be able to return a tuple of array of unique vales and an array of associated indices.
# Nature of the indices depend upon the type of return parameter in the function call.
```

```python
# If we want to know, on average, which type of drive wheel is most valuable, we can group "drive-wheels" 
# and then average them.

# We can select the columns 'drive-wheels', 'body-style' and 'price', then assign it to the variable "df_group_one".
```

```python
df_group_one = df[['drive-wheels','body-style','price']]
```

```python
# grouping results "driver-wheels"
df_group_one = df_group_one.groupby(['drive-wheels'],as_index=False).mean()
df_group_one
```

```python
# "body-style"
df_group_one = df_group_one.groupby(['body-style'],as_index=False).mean()
df_group_one
```

```python
# both "driver-wheels" and "body-style"
df_group_two = df_group_one.groupby(['body-style','drive-wheels'],as_index=False).mean()
df_group_two
```

```python
# This grouped data is much easier to visualize when it is made into a pivot table. A pivot table is like an Excel spreadsheet, 
#with one variable along the column and another along the row. 
# We can convert the dataframe to a pivot table using the method "pivot " to create a pivot table from the groups.
```

```python
grouped_pivot1 = df_group_two.pivot(index='drive-wheels',columns='body-style')
grouped_pivot1
```

```python
# In this case, we will leave the drive-wheel variable as the rows of the table, 
# and pivot body-style to become the columns of the table
```

```python
grouped_pivot1 = grouped_pivot.fillna(0) #fill missing values with 0
grouped_pivot1
```

```python
# Use the "groupby" function to find the average "price" of each car based on "num-of-doors" anf "fuel-system" ?
df_A=df[["num-of-doors","fuel-system" ,"price"]]  # define df columns
df_A=df_A.groupby(["num-of-doors","fuel-system"],as_index=False).mean()  # group "num-of-doors","fuel-system" and the price mean( )
df_A    # call the table 
```

```python
# convert it to a pivot table
Grouped_pivot2=df_A.pivot(index="num-of-doors",columns="fuel-system")
Grouped_pivot2 = Grouped_pivot2.fillna(0) #fill missing values with 0
Grouped_pivot2
```
