```python
df.head(10)
```

```python
import numpy as np

# replace "?" to NaN
df.replace("?", np.nan, inplace = True)
df.head(10)
```

```python
# Evaluating for Missing Data
#The missing values are converted to Python's default. We use Python's built-in functions to identify these missing values.
#There are two methods to detect missing data:

# .isnull()
# .notnull()
# The output is a boolean value indicating whether the value that is passed into the argument is in fact missing data.
# True stands for missing data and False means not missing data
```

```python
missing_data=df.isnull()
missing_data.head()
```

```python
#Count missing values in each column using for loop the method ".value_counts()" counts the number of "True" values.

for column in missing_data.columns.values.tolist(): # numpy.ndarray' object is not callable so need to change it to list
    print(column)   # print all columns in the table
    print (missing_data[column].value_counts())  # print the counts of False or True for each column using .value_counts()
    print("")         # print the name and dtype of each column
```

```python
# How to deal with missing data?
# 1. drop data ( when the whole column or row are empty data)
# a. drop the whole row
# b. drop the whole column
# 2. replace data
# a. replace it by mean
# b. replace it by frequency
# c. replace it based on other functions
```

```python
# Calculate the mean of df in numpy
avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0) # Find the mean of the whole column
                            # a.mean() takes an axis argument axis = 0 for column and axis = 1 for rwo
print("Average of normalized-losses:", avg_norm_loss)
```

```python
#  Replace "NaN" by mean value in "normalized-losses" column

df["normalized-losses"].replace(np.nan,avg_norm_loss,inplace=True)
```

```python
df.head()
```

```python
# Calculate the mean value for 'bore' column
avg_bore= df["bore"].astype("float").mean(axis=0)
print ( "average of normalized-losses:", avg_bore)
```

```python
#  Replace "NaN" by mean value in "bore" column

df["bore"].replace(np.nan,avg_bore,inplace=True)
df.head()  # by default the argument for .head( ) is 5
```

```python
# Write your code below and press Shift+Enter to execute 
avg_stroke=df["stroke"].astype("float").mean(axis=0)
print("stroke:", avg_stroke)
df["stroke"].replace(np.nan,avg_stroke)
df.head()
```

```python
#Calculate the mean value for the 'horsepower' column:
avg_horsepower = df['horsepower'].astype('float').mean(axis=0)
print("Average_horsepower:", avg_horsepower)
df["average horsepower"].replace(np.nan,avg_horsepower)
```

```python
# Calculate the mean value for 'peak-rpm' column:
avg_peakrpm=df['peak-rpm'].astype('float').mean(axis=0)
print("Average peak rpm:", avg_peakrpm)
df['peak-rpm'].replace(np.nan, avg_peakrpm, inplace=True)
```

```python
# To see which values are present in a particular column, we can use the ".value_counts()" method:
# # see how many doors counts
df['num-of-doors'].value_counts()
```

```python
# Normalization is the process of transforming values of several variables into a similar range.
#Typical normalizations include scaling the variable so the variable average is 0, scaling the variable so the variance is 1,
# or scaling variable so the variable values range from 0 to 1
```

```python
#To demonstrate normalization, let's say we want to scale the columns "length", "width" and "height"

#Target:would like to Normalize those variables so their value ranges from 0 to 1.

# Approach: replace original value by (original value)/(maximum value)
```

```python
# replace (original value) by (original value)/(maximum value)
df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
df.head()
```

```python
# replace (original value) by (original value)/(maximum value)
df['height'] = df['height']/df['height'].max()
# show the scaled columns
df[["length","width","height"]].head()  # Show the wanted columns df["C1", "C2"...] .head()
```

```python
# Here we can see, we've normalized "length", "width" and "height" in the range of [0,1].
```

```python
# Binning: Binning is a process of transforming continuous numerical variables into discrete categorical 'bins', for grouped analysis.
#In our dataset, "horsepower" is a real valued variable ranging from 48 to 288,
#it has 57 unique values. What if we only care about the price difference between cars with high horsepower, medium horsepower,
# and little horsepower (3 types)? Can we rearrange them into three ‘bins' to simplify analysis?

# We will use the Pandas method 'cut' to segment the 'horsepower' column into 3 bins
```

```python
#Convert data to correct format
#df["horsepower"]=df["horsepower"].astype(int, copy=True)
```

```python
# Lets plot the histogram of horspower, to see what the distribution of horsepower looks like.

%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(df["horsepower"])

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
# Data Standardization. Standardization refers to shifting the distribution of each attribute to 
# have  a mean of zero and a standard deviation of one (unit variance). 
#It is useful to standardize attributes for a model that relies on the distribution of attributes such as Gaussian processes
```

```python
# We would like 3 bins of equal size bandwidth so 
#we use numpy's linspace(start_value, end_value, numbers_generated function.
```

```python
# We build a bin array, with a minimum value to a maximum value, with bandwidth calculated above. 
# The bins will be values used to determine when one bin ends and another begins.
bins = np.linspace(min(df["horsepower"]), max(df["horsepower"]), 4)
bins

# # df["horsepower"]=df["horsepower"].astype(int, copy=True)
# will cause problem if horsepower is int
# *TypeError: '<' not supported between instances of 'float' and 'str'**  
#if we dont change to the right format for horsepower from "object" to "float"
```

```python
print(min(df["horsepower"]))
```

```python
# set the group name
group_names = ['Low', 'Medium', 'High']
```

```python
# We apply the function "cut" the determine what each value of "df['horsepower']" belongs to.
df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels=group_names , include_lowest=True)
# , include_lowest=True to include the min=48 belongs to low
df[['horsepower','horsepower-binned']].head(20)
```

```python
# See the number of vehicles in each bin.

df["horsepower-binned"].value_counts()  # Use .value_counts()
```

```python
# plot the distribution of each bin.

%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
pyplot.bar(group_names, df["horsepower-binned"].value_counts())

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
# What is an indicator variable?
#An indicator variable (or dummy variable) is a numerical variable used to label categories.
# They are called 'dummies' because the numbers themselves don't have inherent meaning.
```

```python
# Example
# We see the column "fuel-type" has two unique values, "gas" or "diesel". Regression doesn't understand words, only numbers. 
# To use this attribute in regression analysis, we convert "fuel-type" into indicator variables.
```

```python
# We will use the panda's method 'get_dummies' to assign numerical values to different categories of fuel type.
```

```python
df.columns
```

```python
headers
```

```python
# get indicator variables and assign it to data frame "dummy_variable_1"
dummy_variable_1 = pd.get_dummies(df["fuel-type"])
dummy_variable_1.head()
```

```python
# change column names for clarity
dummy_variable_1.rename(columns={'fuel-type-diesel':'gas', 'fuel-type-diesel':'diesel'}, inplace=True)
dummy_variable_1.head()
```

```python
# We now have the value 0 to represent "gas" and 1 to represent "diesel" in the column "fuel-type".
#We will now insert this column back into our original dataset.
# merge data frame "df" and "dummy_variable_1" 
df = pd.concat([df, dummy_variable_1], axis=1)  # axis =1 for column in panda

# drop original column "fuel-type" from "df"
df.drop("fuel-type", axis = 1, inplace=True) # in Panda .drop axis= 1 means work along column
```

```python
df.head()
```

```python
# As above, create indicator variable to the column of "aspiration": "std" to 0, while "turbo" to 1.
dummy_variable_2 = pd.get_dummies(df["aspiration"])
dummy_variable_2.rename(columns={'aspiration':'std', 'aspiration':'turbo'}, inplace=True)
dummy_variable_2.head()
```

```python
# merge data frame "df" and "dummy_variable_2" 
df = pd.concat([df, dummy_variable_2], axis=1)

# drop original column "fuel-type" from "df"
df.drop("aspiration", axis = 1, inplace=True)

df.head()
```
